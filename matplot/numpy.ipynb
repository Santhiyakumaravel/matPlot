```python
#pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([2,4,6,8])
plt.plot(x,y, marker='s',linestyle=':',color='g')
plt.title("simple line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
# plt.show()
# plt.plot(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# for graph the axis scale ploting the graph
x=np.array([1,2,3,4])
y1=np.array([2,4,6,8])
y2=np.array([1,3,5,7])
# plot the axis
# marker is used to connect the points 
plt.plot(x,y1,label='Line1',marker='o',linestyle='--',color='r')
plt.plot(x,y2,label='Line2',marker='s',linestyle=':',color='b')
# marker => o,s,*
# linestyle => '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
# title is to show the graph title
plt.title("Multiple Line Plot")
# label is ued to name
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.show()
# to save the image
plt.savefig("plot1.png")
```

```python
#bar chart
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["apple","banana","orange","pineapple"])
y=np.array([10,24,23,36])
plt.bar(x,y,color='skyblue')
plt.title("Fruit sales")
plt.xlabel("Fruit")
plt.ylabel("Quantity sold")
plt.grid(axis='y')
plt.show()
plt.savefig("plot1.png")
```
